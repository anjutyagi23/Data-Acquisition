Importing all the required packages and modules

```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re
from selenium import webdriver
from functools import reduce
```

Web Scraping for attributes using Selenium (code)

```python
#function to extract data from html format
def extract_data(parameter):
    header=[]
    table = parameter.find_all("table")
    # table
    for j in table:
        if j.get("id") != None and j.get("id") == "grdTableView_DXHeaderTable":
            pattern = re.compile("(grdTableView_tcheader[0-9][0-9]*)")
            header=[str(country.text.strip()) for country in j.find_all('td') if country.get("id") != None and pattern.fullmatch(country.get("id"))]
    header[0]="Countries"
    len(header)
    table = parameter.find_all("table")
    data = []
    data = {country: [] for country in header}
    for j in table:
        if j.get("id") != None and j.get("id") == "grdTableView_DXMainTable":
            for row in j.find_all("tr"):
                cols = [col.text.strip() for col in row.find_all('td')]
                for i in range(len(header)):
                    data[header[i]].append(cols[i])
    df=pd.DataFrame(data)
    return df #converting python dictionary to data frame using PANDAS

#function to get the source page(html format) and parse it to beautifull soup
def scrap_data(url):
    driver = webdriver.Chrome() #calling selenium driver
    driver.get(url)
    
    rows = driver.find_elements_by_css_selector("tr.dxgvDataRow_GridDefaultTheme")
    len(rows)
    li = driver.find_element_by_xpath("//*[@id='grdTableView_DXDataRow49']/td[1]") 
    li.location_once_scrolled_into_view #scroll down
    
    rows1 = driver.find_elements_by_css_selector("tr.dxgvDataRow_GridDefaultTheme")
    len(rows)
    time.sleep(2)
    #get the element reference with find_element()
    li1 = driver.find_element_by_xpath("//*[@id='grdTableView_DXDataRow149']/td[1]") 
    li1.location_once_scrolled_into_view
    
    rows2 = driver.find_elements_by_css_selector("tr.dxgvDataRow_GridDefaultTheme")
    len(rows)
    time.sleep(2)
    #get the element reference with find_element()
    li2 = driver.find_element_by_xpath("//*[@id='grdTableView_DXDataRow199']/td[1]")
    li2.location_once_scrolled_into_view
    
    time.sleep(5)
    response = driver.page_source
    soup = BeautifulSoup(response, "html.parser")
    driver.quit()
    return extract_data(soup)
    
#Unemployment Indicator function call
def unemployment_rate():
    url = "https://databank.worldbank.org/reports.aspx?source=2&series=SL.UEM.TOTL.ZS&country="
    df_unemployment_rate=scrap_data(url)
    df_final_unemployment_rate=pd.melt(df_unemployment_rate, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_unemployment_rate.columns = ['Country','Year','Unemployment Rate']
    return df_final_unemployment_rate

#Literacy Rate Indicator function call
def literacy_rate():
    url = "https://databank.worldbank.org/reports.aspx?source=2&series=SE.ADT.LITR.ZS&country="
    df_literacy_rate=scrap_data(url)
    df_final_literacy_rate=pd.melt(df_literacy_rate, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_literacy_rate.columns = ['Country','Year','Literacy Rate']
    return df_final_literacy_rate

#GDP Growth Rate Indicator function call
def GDP_growth_rate():
    url= "https://databank.worldbank.org/reports.aspx?source=2&series=NY.GDP.MKTP.KD.ZG&country="
    df_GDP_growth_rate=scrap_data(url)
    df_final_GDP_growth_rate=pd.melt(df_GDP_growth_rate, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_GDP_growth_rate.columns = ['Country','Year','GDP Growth Rate']
    return df_final_GDP_growth_rate

#GDP Indicator function call
def GDP():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=NY.GDP.MKTP.CD&country="
    df_GDP=scrap_data(url)
    df_final_GDP=pd.melt(df_GDP, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_GDP.columns = ['Country','Year','GDP']
    return df_final_GDP

#Population Indicator function call
def population():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SP.POP.TOTL&country="
    df_population=scrap_data(url)
    df_final_population=pd.melt(df_population, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_population.columns = ['Country','Year','Population']
    return df_final_population

#External Debt indicator function call
def external_debt():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=DT.DOD.DECT.CD&country="
    df_external_debt=scrap_data(url)
    df_final_external_debt=pd.melt(df_external_debt, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_external_debt.columns = ['Country','Year','External Debt']
    return df_final_external_debt

#Poverty Indicator function call
def poverty_headcount_ratio():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SI.POV.NAHC&country="
    df_poverty_headcount_ratio=scrap_data(url)
    df_final_poverty_headcount_ratio=pd.melt(df_poverty_headcount_ratio, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_poverty_headcount_ratio.columns = ['Country','Year','Poverty Headcount Ratio']
    return df_final_poverty_headcount_ratio

def poverty_gap_1(): #Poverty gap at $1.90 a day
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SI.POV.GAPS&country="
    df_poverty_gap_1=scrap_data(url)
    df_final_poverty_gap_1=pd.melt(df_poverty_gap_1, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_poverty_gap_1.columns = ['Country','Year','Poverty gap at $1.90 a day']
    return df_final_poverty_gap_1

def poverty_gap_2():  #Poverty gap at $3.20 a day (2011 PPP) (%)
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SI.POV.LMIC.GP&country="
    df_poverty_gap_2=scrap_data(url)
    df_final_poverty_gap_2=pd.melt(df_poverty_gap_2, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_poverty_gap_2.columns = ['Country','Year','Poverty gap at $3.20 a day']
    return df_final_poverty_gap_2

def poverty_gap_3(): #Poverty gap at $5.50 a day (2011 PPP) (%) 
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SI.POV.UMIC.GP&country="
    df_poverty_gap_3=scrap_data(url)
    df_final_poverty_gap_3=pd.melt(df_poverty_gap_3, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_poverty_gap_3.columns = ['Country','Year','Poverty gap at $5.50 a day']
    return df_final_poverty_gap_3

#Primary Education Completion Rate Indicator function call
def primary_education_completion_rate(): 
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SE.PRM.CMPT.ZS&country="
    df_primary_education_completion_rate=scrap_data(url)
    df_final_primary_education_completion_rate=pd.melt(df_primary_education_completion_rate, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_primary_education_completion_rate.columns = ['Country','Year','Primary Education Completion Rate']
    return df_final_primary_education_completion_rate

#GNI Per Capita Indicator function call
def GNI_per_capita():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=NY.GNP.PCAP.PP.CD&country="
    df_GNI_per_capita=scrap_data(url)
    df_final_GNI_per_capita=pd.melt(df_GNI_per_capita, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_GNI_per_capita.columns = ['Country','Year','GNI Per Capita']
    return df_final_GNI_per_capita

#Current Health_expenditure Indicator function call
def current_health_expenditure():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=NY.GNP.PCAP.PP.CD&country="
    df_current_health_expenditure=scrap_data(url)
    df_final_current_health_expenditure=pd.melt(df_current_health_expenditure, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_current_health_expenditure.columns = ['Country','Year','Current Health Expenditure']
    return df_final_current_health_expenditure

#Death Rate due to Malnutrition Indicator function call
def death_rate_due_to_malnutrition():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SH.DTH.COMM.ZS&country="
    df_death_rate_due_to_malnutrition=scrap_data(url)
    df_final_death_rate_due_to_malnutrition=pd.melt(df_death_rate_due_to_malnutrition, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_death_rate_due_to_malnutrition.columns = ['Country','Year','Death Rate(Malnutrition)']
    return df_final_death_rate_due_to_malnutrition

#Employment in agriculture Indicator function call
def employment_in_agriculture():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SL.AGR.EMPL.ZS&country="
    df_employment_in_agriculture=scrap_data(url)
    df_final_employment_in_agriculture=pd.melt(df_employment_in_agriculture, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_employment_in_agriculture.columns = ['Country','Year','Employment in Agriculture']
    return df_final_employment_in_agriculture

#Employment in industry Indicator function call
def employment_in_industry():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SL.IND.EMPL.ZS&country="
    df_employment_in_industry=scrap_data(url)
    df_final_employment_in_industry=pd.melt(df_employment_in_industry, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_employment_in_industry.columns = ['Country','Year','Employment in Industry']
    return df_final_employment_in_industry

#Employment in services Indicator function call
def employment_in_services():
    url="https://databank.worldbank.org/reports.aspx?source=2&series=SL.SRV.EMPL.ZS&country="
    df_employment_in_services=scrap_data(url)
    df_final_employment_in_services=pd.melt(df_employment_in_services, id_vars='Countries', value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'])
    df_final_employment_in_services.columns = ['Country','Year','Employment in Services']
    return df_final_employment_in_services
```

Data is converted in Data Frames and comcatenated into one Data Frame

```python
#concatenated different indicator's Data Frames
dfs=[population(),unemployment_rate(), literacy_rate(), GDP_growth_rate(), GDP(),external_debt(), poverty_headcount_ratio(),poverty_gap_1(), poverty_gap_2(), poverty_gap_3(), primary_education_completion_rate(), GNI_per_capita(), current_health_expenditure(), death_rate_due_to_malnutrition(), employment_in_agriculture(), employment_in_industry(), employment_in_services()]
df_final = reduce(lambda left,right: pd.merge(left,right,on=['Country', 'Year'], how='left'), dfs)
df_final
```

Final Data Frame is saved in CSV file

```python
#saved final data frame to CSV file
df_final.to_csv("./data/Poverty-II.csv", index = False)
```
